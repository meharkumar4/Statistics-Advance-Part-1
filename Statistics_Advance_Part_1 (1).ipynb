1. What is a random variable in probability theory.

4. What are probability distribution functions (PDF)?

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
- The Probability Distribution Function (PDF) tells us how the probabilities are distributed at individual points (for discrete variables using PMF) or the density of probability at specific values (for continuous variables using PDF). In contrast, the Cumulative Distribution Function (CDF) gives the total probability up to a certain value. It shows the probability that a random variable is less than or equal to a certain number. While the PDF shows ‚Äúhow likely‚Äù a specific value is, the CDF shows the accumulated probability from the start up to that value. So, the CDF is like a running total of probabilities and always increases from 0 to 1, making it super useful when you want to know the overall chance up to a point, not just at a point.

11. What are the characteristics of a normal distribution/
- It's a **bell-shaped** curve that shows how values are **distributed around the mean** in many real-life situations (like height, test scores, etc.).



###  **Key Characteristics:**

1. **Symmetrical Shape**  
   - The curve is perfectly **symmetric about the mean**  
   - Left and right sides are mirror images

2. **Mean = Median = Mode**  
   - All three are located at the **center** of the distribution

3. **Bell-Shaped Curve**  
   - The curve rises to a peak at the mean and **tapers off** on both sides

4. **Defined by Two Parameters:**  
   - **Œº (mu)** = Mean ‚Üí center of the distribution  
   - **œÉ (sigma)** = Standard deviation ‚Üí spread or width of the curve

5. **68-95-99.7 Rule (Empirical Rule)**  
   - ~68% of data falls within **¬±1œÉ**  
   - ~95% within **¬±2œÉ**  
   - ~99.7% within **¬±3œÉ**

6. **Total Area = 1**  
   - The total area under the curve represents the **entire probability** and equals **1**

7. **Asymptotic**  
   - The tails never touch the x-axis completely; they **extend to infinity**




12. What is the standard normal distribution, and why is it important?
- The **Standard Normal Distribution** is a **special case** of the normal distribution where:

- **Mean (Œº) = 0**  
- **Standard deviation (œÉ) = 1**

It's still the same bell-shaped curve, just **centered at 0** and **measured in standard deviations**.

### **Why is it Important?**

Because it makes **calculations and comparisons easy**!  
Instead of working with different scales (heights, marks, weights), we **convert values into "Z-scores"** using:

\[
Z = \frac{X - \mu}{\sigma}
\]

Where:
- \(X\) = raw score  
- \(\mu\) = mean  
- \(\sigma\) = standard deviation  
- \(Z\) = how many standard deviations \(X\) is from the mean




13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- The Central Limit Theorem (CLT) says:

If you take a large number of random samples from any population (with any shape), the distribution of the sample means will approximate a normal distribution ‚Äî no matter what the original population looks like, as long as the sample size is large enough (usually n ‚â• 30 is good).

### &#x20; **Why is it so Important?**

Because it‚Äôs the **foundation** of most statistical methods. Here's why:

1.  **Enables use of the normal distribution**

   - Even if the data is not normal, the **mean of samples** will be!

2.  **Allows accurate estimation**

   - Lets us calculate **confidence intervals**, **Z-scores**, and do **hypothesis testing** ‚Äî even with non-normal data.

3.  **Supports real-world decision-making**

   - In business, economics, medicine, psychology ‚Äî CLT powers predictions, analysis, and quality control.





14. How does the Central Limit Theorem relate to the normal distribution/
- The Central Limit Theorem explains why the normal distribution is so commonly used in statistics.

Even if the original population is not normally distributed, the distribution of the sample means (i.e., the averages of repeated random samples) will tend to become normal as:

* The sample size increases (typically n ‚â• 30 is good)

* The samples are independent and randomly selected

17. What are point estimates and interval estimates in statistics?
- * A point estimate is a single value that serves as the best guess of a population parameter (like mean or proportion).


* An interval estimate gives a range of values (called a confidence interval) that is likely to contain the true population parameter ‚Äî along with a confidence level (like 95%).



18. What is the significance of confidence intervals in statistical analysis?
- A confidence interval (CI) provides a range of values that‚Äôs likely to include the true population parameter, like a mean or proportion ‚Äî based on your sample data.



###  **Why Are They So Important?**

1.  **They Show Precision**

   - A **narrower interval** = more precise estimate
   - A **wider interval** = more uncertainty

2.  **They Reflect Confidence**

   - A 95% CI means: ‚ÄúIf we repeated this process many times, **95% of the intervals** would capture the true value.‚Äù

3.  **They Go Beyond Point Estimates**

   - Instead of saying ‚Äúthe average is 50,‚Äù you say:\
     ‚ÄúI‚Äôm 95% confident the average is **between 47 and 53**.‚Äù

4.  **They Help Avoid False Certainty**

   - CI shows **margin of error**, not just a single guess
   - That makes decisions **data-driven**, not assumptions-based

5.  **They Aid in Hypothesis Testing**

   - If the CI **does not contain** the hypothesized value (like population mean), it supports **rejecting the null hypothesis**



19. What is the relationship between a Z-score and a confidence interval?

22. What is the concept of expected value in a probability distribution?

\[

- \( P(x_i) \) = probability of that outcome

```python
#3 Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution>
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x : int (0 or 1)
        The outcome (0 or 1).
    p : float (0 <= p <= 1)
        The probability of success (X = 1).

    Returns:
    float
        Probability of observing x in a Bernoulli(p) distribution.
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1")

    return p**x * (1 - p)**(1 - x)

# Example usage
print("P(X=0):", bernoulli_pdf(0, 0.7))  # Output: 0.3
print("P(X=1):", bernoulli_pdf(1, 0.7))  # Output: 0.7
```

```python
#9 Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    """
    Calculate Z-scores for a dataset and plot them.

    Parameters:
    data : list or np.array
        The input dataset.
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    if std == 0:
        raise ValueError("Standard deviation is zero. Z-scores cannot be calculated.")

    z_scores = (data - mean) / std

    # Plotting
    plt.figure(figsize=(10, 5))
    plt.plot(z_scores, marker='o', linestyle='-', color='blue', label='Z-score')
    plt.axhline(0, color='gray', linestyle='--')
    plt.axhline(1, color='green', linestyle='--', label='1 Std Dev')
    plt.axhline(-1, color='green', linestyle='--')
    plt.title('Z-scores of Dataset')
    plt.xlabel('Data Index')
    plt.ylabel('Z-score')
    plt.grid(True)
    plt.legend()
    plt.show()

    return z_scores

# Example usage
sample_data = np.random.normal(50, 10, 100)
z = calculate_and_plot_zscores(sample_data)
```

```python
#18 Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    z = (x - mean) / std
    return z

def plot_standard_normal_with_z(z):
    x_vals = np.linspace(-4, 4, 1000)
    y_vals = norm.pdf(x_vals)

    plt.figure(figsize=(10, 5))
    plt.plot(x_vals, y_vals, label='Standard Normal PDF', color='blue')
    plt.fill_between(x_vals, y_vals, where=(x_vals <= z), color='orange', alpha=0.5, label=f'Z ‚â§ {z:.2f}')
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
    plt.title('Z-score Compared to Standard Normal Distribution')
    plt.xlabel('Z-value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example values
data_point = 115
mean = 100
std = 15

# Step 1: Calculate Z-score
z_score = calculate_z_score(data_point, mean, std)
print(f"Z-score for X = {data_point}: {z_score:.2f}")

# Step 2: Plot comparison with standard normal distribution
plot_standard_normal_with_z(z_score)
```

```python
#19  Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    sample_size = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Step 1: Calculate Z-statistic
    z = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Step 2: Calculate p-value (two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(z)))

    # Step 3: Make a decision
    reject_null = p_value < alpha

    # Print results
    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Z-statistic = {z:.4f}")
    print(f"P-value = {p_value:.4f}")
    if reject_null:
        print("üö® Reject the null hypothesis (H‚ÇÄ). Significant difference found.")
    else:
        print("‚úÖ Fail to reject the null hypothesis (H‚ÇÄ). No significant difference.")

# Example usage
np.random.seed(42)
sample = np.random.normal(loc=102, scale=15, size=50)  # sample data with mean around 102
population_mean = 100
population_std = 15

z_test(sample, population_mean, population_std)
```

```python
#20 Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation
    n = len(data)

    z_score = norm.ppf((1 + confidence) / 2)
    margin_of_error = z_score * (sample_std / np.sqrt(n))

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
    return (lower_bound, upper_bound)

# Example usage
np.random.seed(1)
sample_data = np.random.normal(loc=100, scale=15, size=50)
confidence_interval(sample_data)
```

```python
#21 Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
from scipy.stats import norm

def generate_normal_data_and_ci(mean=100, std_dev=15, size=50, confidence=0.95):
    # Step 1: Generate data
    np.random.seed(42)
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Step 2: Calculate sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)

    # Step 3: Compute margin of error
    z_score = norm.ppf((1 + confidence) / 2)
    margin_of_error = z_score * (sample_std / np.sqrt(size))

    # Step 4: Confidence interval
    lower = sample_mean - margin_of_error
    upper = sample_mean + margin_of_error

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    return data, (lower, upper)

# Example usage
data, ci = generate_normal_data_and_ci()
```

```python
#23  Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
lambda_val = 5  # average number of events
x = np.arange(0, 15)  # possible values of k

# Calculate CDF values
cdf = poisson.cdf(x, mu=lambda_val)

# Print CDF for a specific value
k = 7
print(f"P(X ‚â§ {k}) when Œª = {lambda_val} is: {poisson.cdf(k, mu=lambda_val):.4f}")

# Plotting the CDF
plt.figure(figsize=(10, 5))
plt.step(x, cdf, where='post', label='Poisson CDF', color='purple')
plt.scatter(x, cdf, color='black')
plt.title(f'Cumulative Distribution Function of Poisson(Œª = {lambda_val})')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#24 Simulate a random variable using a continuous uniform distribution and calculate its expected value
import numpy as np

# Parameters of the uniform distribution
a = 10   # lower bound
b = 30   # upper bound
n = 10000  # number of samples

# Simulate random variable
samples = np.random.uniform(low=a, high=b, size=n)

# Calculate sample mean (empirical expected value)
sample_mean = np.mean(samples)

# Theoretical expected value
expected_value = (a + b) / 2

# Results
print(f"Theoretical Expected Value: {expected_value:.2f}")
print(f"Empirical Mean from {n} samples: {sample_mean:.2f}")
```

```python
#26 Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution

import numpy as np

# Step 1: Generate data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=1000)

# Step 2: Calculate Range
range_val = np.max(data) - np.min(data)

# Step 3: Calculate IQR
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

# Step 4: Display results
print(f"Range: {range_val:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
print(f"Q1 (25th percentile): {q1:.2f}")
print(f"Q3 (75th percentile): {q3:.2f}")
```

```python
#27 Implement Z-score normalization on a dataset and visualize its transformation
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

# Step 2: Z-score normalization
mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std

# Step 3: Print a few results
print(f"Original Mean: {mean:.2f}, Original Std Dev: {std:.2f}")
print(f"Normalized Mean: {np.mean(z_scores):.2f}, Std Dev: {np.std(z_scores):.2f}")

# Step 4: Visualization
plt.figure(figsize=(12, 5))

# Original data histogram
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

# Z-score normalized data histogram
plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, color='salmon', edgecolor='black')
plt.title('Z-score Normalized Data')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#28 Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution
import numpy as np
from scipy.stats import skew, kurtosis

def analyze_distribution_skew_kurt(mean=0, std_dev=1, size=1000):
    # Step 1: Generate data
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Step 2: Calculate skewness and kurtosis
    skew_val = skew(data)
    kurt_val = kurtosis(data, fisher=False)  # Fisher=False gives Pearson kurtosis (normal = 3)

    # Step 3: Display results
    print(f"Skewness: {skew_val:.4f}")
    print(f"Kurtosis: {kurt_val:.4f}")
    return skew_val, kurt_val

# Example usage
analyze_distribution_skew_kurt()
```
